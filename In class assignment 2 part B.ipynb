```python
print("\n\n###################\n 1- Stripping white space \n###################\n")
# 1- Stripping white space
print(text.strip())


print("\n\n###################\n 2- Remove periods\n###################\n")
# 2- Remove periods
print(text.replace('.',''))

print("\n\n###################\n  3- capitalize text\n###################\n")

# 3- capitalize text
print(text.strip().capitalize())
print("To make this works, i should strip white spaces, and or apply it directly to sentences")

print("\n\n###################\n  4- replace letters with X\n###################\n")
# 4- replace letters with X
text2=""
for c in text:
    text2+= "x" if c.isalpha() else c 
print(text2)


print("\n\n###################\n 5- remove punctuation \n###################\n")
# 5- remove punctuation
text2=""
for c in text:
    text2+= "" if not c.isalpha() else c 
print(text2)


print("\n\n###################\n  6- remove stopwords\n###################\n")
# 6- remove stopwords
list_nostopwords=[]
for token in tokenized_words:
    if token not in stopwords:
        list_nostopwords.append(token)
print(" ".join(list_nostopwords))


print("\n\n###################\n 7- Split sentences \n###################\n")
# 7- Split sentences
_sents = sent_tokenize(text)
print(_sents)


print("\n\n###################\n 8- Tokenize words, 2 approaches \n###################\n")
# 8- Tokenize words, 2 approaches
print("\nTokenize words, app1:",text.split())
print("\nTokenize words, app2:",tokenized_words)


print("\n\n###################\n  9- stemming\n###################\n")
# 9- stemming

ps = PorterStemmer()
for w in list_nostopwords:
    print(w," => ",ps.stem(w))


print("\n\n###################\n  10- Apply stemming on words \"universal\" and \"university\"\n###################\n")
# 10- Apply stemming on words "universal" and "university" and see the result. then comment on drawback of stemming
print("Universal:",ps.stem('universal'))
print("Universe:",ps.stem('universe'))
print("Using this method will not recognize the POS of the word, \n and if a word is derived from another, using th same root, just as in the example above.")


```
