# 2-Explore inner-join, outer-join, left-join, and right-join of tables in pandas

```python
import pandas as pd

# Sample DataFrame 1
data1 = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 22, 28, 35]
}
df1 = pd.DataFrame(data1)

# Sample DataFrame 2
data2 = {
    'ID': [4, 5, 6, 7, 8],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Seattle'],
    'Salary': [80000, 75000, 60000, 90000, 70000]
}
df2 = pd.DataFrame(data2)
```

```python
inner_join_result = pd.merge(df1, df2, on='ID', how='inner')
print(inner_join_result)
```

```python
outer_join_result = pd.merge(df1, df2, on='ID', how='outer')
print(outer_join_result)
```

```python
left_join_result = pd.merge(df1, df2, on='ID', how='left')
print(left_join_result)
```

```python
right_join_result = pd.merge(df1, df2, on='ID', how='right')
print(right_join_result)
```

```python
# This was on the test
test_join_result = pd.merge(df1, df2)
print(test_join_result)
```

## 3- Explore the application of these functions in Python with some examples:  
    apply applymap map
    
     However, remember that apply works along rows or columns, whereas applymap works element-wise on DataFrames. For Series objects, you should use apply to operate along the axis. For DataFrames, use apply when you need to apply a function along rows or columns and applymap when you want to apply a function element-wise to each individual element of the DataFrame.

#### APPLY

```python
import pandas as pd

# Sample DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)

# Define a function to square the values
def square(x):
    return x**2

# Apply the 'square' function to each element in the DataFrame
df_square = df.apply(square)
print(df_square)
```

#### APPLYMAP

```python
import pandas as pd

# Sample DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)

# Define a function to square the values
def square(x):
    return x**2

# Apply the 'square' function to each element in the DataFrame
df_square = df.applymap(square)
print(df_square)
```

#### MAP

```python
import pandas as pd

# Sample Series
data = ['one', 'two', 'three', 'two', 'one']
s = pd.Series(data)

# Define a mapping dictionary
mapping = {
    'one': 1,
    'two': 2,
    'three': 3
}

# Use the 'map' function to replace values in the Series
s_mapped = s.map(mapping)
print(s_mapped)
```

```python
# Mean
nums= [2,6,6,13,33,22,1,8,23,5,7,8,5,7,6,3,7,8,8]
print("Mean is {}".format(sum(nums)//len(nums)))
```

```python
nums.sort()
print("Median from {}\n is {}".format(nums, nums[len(nums)//2]))
```

```python
mode=0
mode_count=0
for n in nums:
    if nums.count(n) >= mode_count:
        mode_count=nums.count(n)
        mode=n
print("Mode in {}\nis: {} with {} appearances".format(nums, mode, mode_count))
```


## Standard Deviation
is a measure of variation or dispertsion of a dataset
it is a number used to tell how measurements for a group are spread out from the average/mean or expected value


```python
students = [2,4,4,4,5,5,7,9]
mean = sum(students)//len(students)
mean
```

```python
import numpy as np
```

```python
def get_std(nums):
    mean = sum(nums)//len(nums)
    std=0
    for n in nums:
        std+= (n-mean)**2
    #avg 
    std=std//len(nums)
    #sqUARE ROOT
    std = std**0.5
    return std
std=get_std(students)
std_np = np.std(students)
print("my std is {}, np.std is: {}".format(std, std_np))
```
