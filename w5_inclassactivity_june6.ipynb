    Ugly number are positive numbers whose only prime factors are 2,3,5
    the sequence 12345678910 12
    shows the first 10 ugly numbers
    
    note 1, is typically treated as ugly number

```python
def is_ugly(num):
    print(num,end=" ")
    if num == 1:
        return True
    if(num%4==0):
        return False
    for i in range(6,num):
        if(num%i==0 and (num%5==0 and num%3==0 and num%2==0)):
            return False
    if(num%5==0 or num%3==0 or num%2==0):
        return True
    return False
print(is_ugly(11))  # False (Prime factor other than 2, 3, or 5)
print(is_ugly(1))   # True (Special case where 1 is considered an ugly number)
print(is_ugly(5))   # True (Only has prime factors 5)
print(is_ugly(15))  # True (Has prime factors 3 and 5)
print(is_ugly(20))  # True (Has prime factors 2 and 5)
print(is_ugly(16))  # True (Has prime factor 2)
print(is_ugly(12))  # True (Has prime factors 2 and 3)
print(is_ugly(121))  # False (Has a prime factor other than 2, 3, or 5)
print(is_ugly(21))   # False (Has a prime factor other than 2, 3, or 5) <<<<<<
print(is_ugly(51))   # False (Has a prime factor other than 2, 3, or 5) <<<<<<
print(is_ugly(156))  # False (Has a prime factor other than 2, 3, or 5)
print(is_ugly(23))   # False (Has a prime factor other than 2, 3, or 5)
print(is_ugly(17))   # False (Has a prime factor other than 2, 3, or 5)
print(is_ugly(19))   # False (Has a prime factor other than 2, 3, or 5)
```

```python
def get_ugly_number(n):
    ugly = [0] * n  # Create an empty list to store ugly numbers
    ugly[0] = 1  # Set the first ugly number as 1
    next_multiple_of_2 = 2
    next_multiple_of_3 = 3
    next_multiple_of_5 = 5
    i2 = i3 = i5 = 0

    for i in range(1, n):
        # Find the minimum of the multiples of 2, 3, and 5
        ugly[i] = min(next_multiple_of_2, next_multiple_of_3, next_multiple_of_5)

        # Update the indices for the next multiples
        if ugly[i] == next_multiple_of_2:
            i2 += 1
            next_multiple_of_2 = ugly[i2] * 2
        if ugly[i] == next_multiple_of_3:
            i3 += 1
            next_multiple_of_3 = ugly[i3] * 3
        if ugly[i] == next_multiple_of_5:
            i5 += 1
            next_multiple_of_5 = ugly[i5] * 5

    return ugly[-1]  # Return the last element of the ugly numbers list


# Example usage:
n = 12
ugly_number = get_ugly_number(n)
print(f"The {n}th ugly number is: {ugly_number}")
```

```python
def is_ugly(number):
    if number <= 0:
        return False
    
    while number % 2 == 0:
        number = number // 2
    
    while number % 3 == 0:
        number = number // 3
    
    while number % 5 == 0:
        number = number // 5
    
    return number == 1


# Example usage:
num = 30
if is_ugly(num):
    print(f"{num} is an ugly number.")
else:
    print(f"{num} is not an ugly number.")
```
